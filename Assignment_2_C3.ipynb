```python
player_team=[]
player_goals=dict()
mat_played=dict()
assists=dict()
avg=dict()
highest_goals=dict()
tackles=dict()
red=dict()
yellow=dict()
hat_trick=[]
saves=[]

f=open("Uefa.csv","r")
data=f.readline()
```

```python
while(True):

  data=f.readline()
  if not data:
    break;

  print(data)
  #data=data.replace("\n","")
  temp=data.split(",")
  player_team.append(temp[2])
  player_goals.update({temp[1]:temp[4]})
  mat_played.update({temp[1]:temp[3]})
  assists.update({temp[1]:temp[5]})
  avg.update({temp[1]:temp[8]})
  highest_goals.update({temp[1]:int(temp[9])})
  tackles.update({temp[1]:temp[7]})
  red.update({temp[1]:temp[11]})
  yellow.update({temp[1]:temp[12]})
  hat_trick.append(int(temp[13]))
  saves.append(int(temp[6]))


player_team=tuple(player_team)
```

1) Team having Most No of Players. 

```python
print(player_team,"\n\n\n") 

frequency ={}
for team in player_team:
  if team in frequency:
    frequency[team]+=1
  else:
    frequency[team]=1
print(frequency)
marklist = sorted(frequency.items(), key=lambda x:x[1],reverse=True)
sortdict = dict(marklist)
print(sortdict)
print("\n\nTeam which have most players in list is", list(sortdict.keys())[0],"with",list(sortdict.values())[0],"players")
```

2)Players who have scored more than 3 goals.

```python
print(player_goals) 
print("\n\n\nPlayers with more than 3 goals are")
for player in player_goals:
  if player_goals[player]>"10":
     print(player,"\t\t:",player_goals[player])
```

3)Players who played minimum matches in the league

```python

print("\n\n\nPlayers played minimum'",max(mat_played.values()),"'times are:",end=" ")

for player in mat_played:
  if mat_played[player]==max(mat_played.values()):
    print(player,end=", ")
```

4)Most Assists

```python
print(assists)
print(max(assists.values()))

for player in assists:
  if assists[player]==max(assists.values()):
    print("\n\n\n",player,end=" ")


print("assisted most",max(assists.values()),"times")
```

5)Which player has Best avg

```python
print(avg)

for player in avg:
  if avg[player]==max(avg.values()):
    print("\n\n\n",player,end=" ")


print("has best avg of",max(avg.values()))
```

6)Who has most tackles

```python
print(tackles)

for player in tackles:
  if tackles[player]==max(tackles.values()):
    print("\n\n\n",player,end=" ")


print("has tackled most",max(tackles.values()))
```

7)How many players got more than 2 red cards

```python
print(red)
count = 0
for player in red:
  if red[player] > '2':
    print(player,"\t\t:",red[player])
    count=count + 1

print("\n\nTotal",count,"player(s) have more than 2 red cards ") 
```

8)Who have more than 5 yellow cards

```python
print(yellow)
count = 0
for player in yellow:
  if yellow[player] > '5':
    print(player,"\t\t:",yellow[player])
    count=count + 1

print("\n\nTotal",count,"player(s) have more than 5 yellow cards ") 
```

9)How many total hat_tricks scored?

```python
print(hat_trick)


print("\n\nTotal",sum(hat_trick),"hat-tricks scored")  
```

```python
print(saves)


print("\n\nTotal",sum(saves),"saves in the league") 
```
