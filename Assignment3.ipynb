```python
import numpy as np
```

```python
array1=np.array([[23,25,30],[34,78,67],[56,18,34]])
array1
```

```python
array2=np.array([[26,89,45],[56,76,64],[34,98,48]])
array2
```

1.1-Addition

```python
array3=array1 + array2
print("\nUsing Operator:\n",array3)
array3=np.add(array1,array2)
print("\nUsing NumPy Function:\n",array3)
```

1.2-Subtraction

```python
array3=array1 - array2
print("\nUsing Operator:\n",array3)
array3=np.subtract(array1,array2)
print("\nUsing NumPy Function:\n",array3)
```

1.3-Multiplication

```python
array3=array1 * array2
print("\nUsing Operator:\n",array3)
array3=np.multiply(array1,array2)
print("\nUsing NumPy Function:\n",array3)
```

1.4-Division

```python
array3=array1 / array2
print("\nUsing Operator:\n",array3)
array3=np.divide(array1,array2)
print("\nUsing NumPy Function:\n",array3)
```

1.5-Mod

```python
array3=array1 % array2
print("\nUsing Operator:\n",array3)
array3=np.mod(array1,array2)
print("\nUsing NumPy Function:\n",array3)
```

1.6-Dot Product

```python
array3=np.dot(array1,array2)
print(array3)
```

1.7-Transpose

```python
array3=np.transpose(array1)
print(array3)
print("\n")
array3=array1.transpose()
print(array3)
```

2.1-Horizontal Stacking

```python
array3=np.hstack((array1,array2))
array3
```

2.2-Vertical Stacking

```python
array3=np.vstack((array1,array2))
array3
```

2.3-Stacking

```python
array3=np.stack((array1,array2),axis=0)
array3
```

```python
array3=np.stack((array1,array2),axis=1)
array3
```

```python
array3=np.stack((array1,array2),axis=2)
array3
```

1-D Stacking

```python
arr1=np.array([1,5,7])
arr2=np.array([4,9,8])
arr3=np.stack((arr1,arr2),axis=0)
print(arr3)
print("\n")
arr3=np.stack((arr1,arr2),axis=1)
print(arr3)
```
