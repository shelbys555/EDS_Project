```python
 evenlist=[]
oddlist=[]
n=int(input("Enter the range: "))
for x in range(0,n):
  a=int(input())
  if(a%2==0):
     evenlist.append(a)
  else:
     oddlist.append(a)  

print("Even list is",evenlist)
print("Odd list is",oddlist)
```

```python
div_by_3=[]
div_by_5=[]
div_by_10=[]
n=int(input("Enter the range: "))
for x in range(0,n):
  a=int(input())
  if(a%3==0):
    div_by_3.append(a)
  if(a%5==0):
    div_by_5.append(a)
  if(a%10==0):
    div_by_10.append(a)

print("Numbers divisible by 3:",div_by_3)
print("Numbers divisible by 5:",div_by_5)
print("Numbers divisible by 10:",div_by_10)  
```

```python
def fact(a):
  if(a==0):
    return 1
  fact=1
  for x in range(a,0,-1):
    fact=fact*x
  return fact

n=int(input("Enter the number: "))
print("Factorial of the number is",fact(n))
```

```python
def fact(n):
  if(n==0)or(n==1):
    return 1
  else:
    return n*fact(n-1)

a=int(input("Enter the number: "))
print("Factorial of the  number is",fact(a))
```
