```python
f=open("emp.csv","r")

contents=f.read()

lines=contents.split("\n")

eid=[]
name=[]
des=[]
sal=[]

for l in lines:
  words=l.split(",")
  print(words)
  eid.append(int(words[0]))
  name.append(words[1])
  des.append(words[0])
  sal.append(int(words[3]))

print("Employee Id",eid)
print("Employee Name",name)
print("Employee Designation",des)
print("Employee Salary",sal)

print("Maximum Salary",max(sal))

print("Minimum Salary",min(sal))

print("Average Salary",(sum(sal)/len(sal)))


print("Employee with maximum salary:",name[sal.index(max(sal))])

print("Employee with minimum salary:",name[sal.index(min(sal))])

print("Sr managers are",end=" ")
for i in range(len(des)):
  if des[i]=='Sr. Manager':
    print(name[i],end=" ")

print("Employee having salary 80000:",end=" ")
for i in range(len(sal)):
  if sal[i]==80000:
    print(name[i],end=" ")

print("Employee having salary 70000:",end=" ")
for i in range(len(sal)):
  if sal[i]==70000:
    print(name[i])
  else:
    print("No Employee has salary 70000")      
```
