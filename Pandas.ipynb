```python
import pandas as pd

data = pd.read_csv('/content/purchase_data (1).csv')

print(data.to_string())
#find NAN
nan_df = data[data.isnull().any(axis=1)]

data = data.dropna(how='all')
data.head(70)
```

```python
data = data[data['Order Date'].str[0:2]!='Or']

print(data['Order Date'])
```

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])

data['Quantity Ordered']
```

```python
data['Month']= data['Order Date'].str[0:2]
data['Month']= data['Month'].astype('int32')
data.head()
```

```python
data['Month 2']=pd.to_datetime(data['Order Date']).dt.month
data.head()
```

```python
data['Day']=pd.to_datetime(data['Order Date']).dt.day
data.head()
```

```python
data['Year']=pd.to_datetime(data['Order Date']).dt.year
data.head()
```

```python
data['Hours']=pd.to_datetime(data['Order Date']).dt.hour
data.head()
```

```python
data['Minutes']=pd.to_datetime(data['Order Date']).dt.minute
data.head()
```

```python
def get_city(address):
  return address.split(',')[1].strip(" ")

def get_state(address):
  return address.split(',')[2].split(" ")[1]

data['City']= data['Purchase Address'].apply(lambda x: f'{get_city(x)}  ({get_state(x)})')
data.head()
```

```python
data['Sales']  = data['Quantity Ordered'].astype('int') * data['Price Each'].astype('float')
```

```python
data.groupby(['Month']).sum()
```

```python
city_max = data.groupby(['City']).sum()
print(max(city_max))
print(city_max[0])
```
